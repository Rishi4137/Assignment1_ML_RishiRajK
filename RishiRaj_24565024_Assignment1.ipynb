```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv("Crime_Data_from_2020_to_present.csv")
```

```python
dataset.head(10)
```

```python
dataset.info()
```

```python
dataset.drop(columns=['Crm Cd 4','Crm Cd 2','Crm Cd 3'],inplace=True)
```

```python
dataset.info()
```

```python
dataset['AREA NAME'].value_counts().plot(kind='bar')
plt.xlabel('Area Name')
plt.ylabel('Count of Crimes')
# Central is the place where largest no of crimes occur
```

```python
dataset["Crm Cd"].value_counts().head(30).plot(kind='bar')
```

```python
# Crime with code 510(Vehicle theft is the most common crime)
filtered_data = dataset[dataset['Crm Cd'] == 510]
crime_description = filtered_data['Crm Cd Desc'].iloc[0]
print(crime_description)
```

```python
dataset["Crm Cd Desc"].value_counts().head(20).plot(kind='pie')
```

```python
dataset['TIME OCC'].plot(kind='hist', bins=24,edgecolor='black')
plt.show()
#Most crimes Occur between 1-2 pm ,also there is large no of crime occuring between 5-6 pm.
```

```python
dataset['Vict Sex'].value_counts()
```

```python
dataset['Vict Sex'].value_counts().plot(kind='bar')
# Most of victims are males followed by females.
```

```python
mean=dataset['Vict Age'].mean()
print("Mean age of victims:",mean)
max=dataset['Vict Age'].max()
print("Max age of victims:",max)
min=dataset['Vict Age'].min()
print("Min age of victims:",min)
```

```python
dataset.loc[dataset['Vict Age'] < 0, 'Vict Age'] = mean
```

```python
mean=dataset['Vict Age'].mean()
print("Mean age of victims:",mean)
max=dataset['Vict Age'].max()
print("Max age of victims:",max)
min=dataset['Vict Age'].min()
print("Min age of victims:",min)
```

```python
dataset['Vict Age'].plot(kind='box')
```

```python
dataset['Vict Age'].plot(kind='hist',bins=24,title="Age of Victims",edgecolor='black')
#Most of victims are kids between age 0-5
```

```python
dataset['Vict Descent'].value_counts().plot(kind='bar')
#Most of the victimes belongs to hispanic background.
```

```python
dataset['Weapon Desc'].value_counts().head(10).plot(kind='bar',title='Top 10 Weapon Used')
```

```python
dataset['Status'].value_counts()
```

```python
dataset['Status Desc'].value_counts().plot(kind='bar',title='Frequency of Status Descriptions')
plt.xticks(rotation=60)
plt.show()
```

```python
dataset['Premis Desc'].value_counts().head(10).plot(kind='bar')
# Most of the Crime occur in some Streets.
```

```python
dataset['Part 1-2'].value_counts().plot(kind='bar')
# Part 1 crimes are large in number which are more serious crimes,typically involving violence.
```

```python
numerical_cols = dataset.select_dtypes(include='number').columns.tolist()
correlation_matrix = dataset[numerical_cols].corr()
print(correlation_matrix)
```

```python
top_20_codes = dataset['Crm Cd Desc'].value_counts().head(10).index.tolist()
filtered_dataset = dataset[dataset['Crm Cd Desc'].isin(top_20_codes)]
filtered_dataset.plot.scatter(x='Vict Age', y='Crm Cd Desc',)
plt.title('Victim Age vs Top 10 Crimes')
plt.show()
```

```python
sns.violinplot(x='Crm Cd', y='Vict Age', data=filtered_dataset)
plt.xticks(rotation=45)  
plt.show()
```

```python
top_10_codes = dataset['Crm Cd'].value_counts().head(10).index.tolist()
filtered_dataset = dataset[dataset['Crm Cd'].isin(top_10_codes)]
sns.countplot(x='Vict Sex', hue='Crm Cd', data=filtered_dataset)
plt.title("Distribution of Victim Sex by Top 10 Crime Codes")
```

```python
top_20_codes = dataset['Premis Cd'].value_counts().head(10).index.tolist()
filtered_dataset = dataset[dataset['Premis Cd'].isin(top_20_codes)]
sns.countplot(x='Vict Sex', hue='Premis Cd', data=filtered_dataset)
plt.title('Distribution of Victim Sex by Top 10 Premise Types')

```

```python
sns.boxplot(x='Vict Sex', y='Vict Age', data=filtered_dataset, palette='rainbow')
```

```python
sns.boxplot(x='Vict Sex', y='Premis Cd', data=filtered_dataset, palette='rainbow')
```
